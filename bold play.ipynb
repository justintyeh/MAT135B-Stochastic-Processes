```python
import matplotlib.pyplot as plt
import numpy as np

# bold play probabilities
# for p=[0.1 0.25 9/19 0.5]
probabilities = [0.05 + 0.05 * i for i in range(10)]

for p in probabilities:
  Pn = [0]
  Pn[0] = 1
  sizePn = 1
  for nop in range(0,11):
    sizeQn=2*sizePn
    Qn= [0] * (sizeQn + 1)
    #print(sizeQn)
    for i in range(0, sizeQn+1):
      #print(i)
      if i % 2 == 1:
        Qn[i] = Pn[i//2]
      elif i <= sizePn:
        Qn[i] = p * Pn[i]
      else:
        #print(i, len(Qn), i-sizePn, Pn)
        Qn[i] = p + (1-p) * Pn[i-sizePn-1]
    Pn = Qn
    sizePn = sizeQn
  t = np.asarray([i for i in range(sizePn+1)]) / sizePn
  plt.plot(t, Pn)

plt.rcParams['figure.figsize'] = [20, 10]
plt.title('Bold play probabilities')
plt.xlabel('x')
plt.ylabel("P(x)")
```
