```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# gambler's ruin probabilities
n = 100
probabilities = [ 0.2 + .01 * i for i in range(1,31)]
i_n = np.linspace(0, n, num=30) 

Pn = [0] * n
for i, p in enumerate(probabilities):
  if p == 1/2:
    Pn = i_n/n
  else:
    Pn=((1-((1-p)/p)**i_n))/(1-((1-p)/p)**n); 
  plt.plot(i_n, Pn)
plt.axhline(y = 0.5, linestyle = "--", label = "50%")
plt.title("Gambler's Ruin/Random Walk Probabilities")
plt.xlabel("Position i")
plt.ylabel("P_i")
plt.legend(loc = 'best')
plt.rcParams['figure.figsize'] = [20, 10]
plt.show()
```
