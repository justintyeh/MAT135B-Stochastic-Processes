# Longest Consecutive Run of Heads

Toss a fair coin $n$ times. Let $R_n$ = the longest consecutive run of Heads 

ie. $HHHTHHHTTTHHT$, $R_n=3$

Since the graph will look roughly logarithmic, the simulation shows that $\frac{R_n}{log_2n} \rightarrow 1 \text{ as } n → ∞$

**NOTE:** the code may take a while to run since n = 1 billion. You can try running it with smaller values first.

```python
import numpy as np
import matplotlib.pyplot as plt
p = 0.5
n = 10000000

s = [0] * n
sr = [0] * n

x = np.random.uniform(0,1,size=n)
s = (x <= p)
#s3 = np.random.randint(0,2, size=n) # generate random integers [0,1]

cr = 0
for i in range(1, n):
  if s[i] == False:
    cr = 0
  else:
    cr += 1
  sr[i] = cr

t = np.linspace(0,n, num=n)
plt.plot(t, np.maximum.accumulate(sr))
plt.title('Run Size')
plt.xlabel(f'Number of Coin Tosses n = {n}')
plt.ylabel('Longest run of Consecutive Heads')
plt.rcParams['figure.figsize'] = [15, 10]
```
